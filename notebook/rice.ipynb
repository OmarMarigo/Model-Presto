```python
import numpy as np
import os 
import pandas as pd
import geopandas as gpd
```

```python
# Définir le chemin du répertoire contenant les fichiers .geojson
data_rice_sn_2023 = '../../DATA GEOJSON/'


list_data = []

for filename in os.listdir(data_rice_sn_2023):

    if filename.endswith('.geojson'):
     
        chemin_fichier = os.path.join(data_rice_sn_2023, filename)

        data = gpd.read_file(chemin_fichier)
    
        list_data.append(data)
        
combined_data = gpd.GeoDataFrame(pd.concat(list_data, ignore_index=True))

combined_data["class"] = pd.to_numeric(combined_data['class'], errors='coerce')
```

```python
data_rice_01 = gpd.read_file('../../DATA GEOJSON/rice.geojson')
data_rice_02 = gpd.read_file('../../DATA GEOJSON/rice_2.geojson')
```

```python
data_rice = pd.concat([data_rice_01, data_rice_02])
data_rice["Renseigner la culture"] = "riz"
```

```python
other_1 = gpd.read_file('../../DATA GEOJSON/mais.geojson')
other_2 = gpd.read_file('../../DATA GEOJSON/mil.geojson')
other_3 = gpd.read_file('../../DATA GEOJSON/coton.geojson')
other_4 =gpd.read_file('../../DATA GEOJSON/cann_a_sucre.geojson')
other_5 =gpd.read_file('../../DATA GEOJSON/archide_1.geojson')

data_other_2 = pd.concat([other_1, other_2, other_3, other_4, other_5])
data_other_2["Renseigner la culture"]= "other_2"

data_other_2.shape
```

```python
other7 = gpd.read_file('../../DATA GEOJSON/rice_others_st_louis.geojson')
other7["Renseigner la culture"] = np.where(other7["class"]== 0 , 'riz', 'other_1')
other7
```

```python
other1 = gpd.read_file('../../DATA GEOJSON/build.geojson')
other2 = gpd.read_file('../../DATA GEOJSON/others.geojson')
other3 = gpd.read_file('../../DATA GEOJSON/others2.geojson')
other4 = gpd.read_file('../../DATA GEOJSON/ble.geojson')
other5 = gpd.read_file('../../DATA GEOJSON/soil.geojson')
other6 = gpd.read_file('../../DATA GEOJSON/water.geojson')

data_other_1 = pd.concat([other1,other2, other3, other4, other5,other6])
data_other_1["Renseigner la culture"]= "other_1"

data_other_1.shape
```

```python
rice_data_modele = pd.concat([data_other_1, data_rice, data_other_2, other7])
rice_data_modele
```

```python
def convert_multipolygon_to_polygon(geometry):
    if geometry.geom_type == 'Polygon':
        return geometry  
    elif geometry.geom_type == 'MultiPolygon':
        return geometry.convex_hull  
    elif geometry.geom_type == 'Point':
        return geometry.buffer(0.1) 
    return geometry  #
# Appliquer la fonction à chaque géométrie de la colonne 'geometry'
rice_data_modele['geometry'] = rice_data_modele['geometry'].apply(convert_multipolygon_to_polygon)

# Afficher le résultat
print(rice_data_modele)
```
