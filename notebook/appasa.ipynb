```python
def count_values_in_range(df, column, range_min, range_max):
   
    filtered_df = df[(df[column] > range_min) & (df[column] <= range_max)]
    return filtered_df[column].value_counts().sum()


sum_0_to_20 = count_values_in_range(df, 'rh98', 0, 20)
sum_20_to_40 = count_values_in_range(df, 'rh98', 20, 40)
sum_40_to_60 = count_values_in_range(df, 'rh98', 40, 60)
sum_60_to_80 = count_values_in_range(df, 'rh98', 60, 80)
sum_80_to_100 = count_values_in_range(df, 'rh98', 80, 100)

print("Somme des occurrences pour la plage 0-20:", sum_0_to_20)
print("Somme des occurrences pour la plage 20-40:", sum_20_to_40)
print("Somme des occurrences pour la plage 40-60:", sum_40_to_60)
print("Somme des occurrences pour la plage 60-80:", sum_60_to_80)
print("Somme des occurrences pour la plage 80-100:", sum_80_to_100)
```

```python

sum_0_to_20 = count_values_in_range(df, 'rh100', 0, 20)
sum_20_to_40 = count_values_in_range(df, 'rh100', 20, 40)
sum_40_to_60 = count_values_in_range(df, 'rh100', 40, 60)
sum_60_to_80 = count_values_in_range(df, 'rh100', 60, 80)
sum_80_to_100 = count_values_in_range(df, 'rh100', 80, 100)

print("Somme des occurrences pour la plage 0-20:", sum_0_to_20)
print("Somme des occurrences pour la plage 20-40:", sum_20_to_40)
print("Somme des occurrences pour la plage 40-60:", sum_40_to_60)
print("Somme des occurrences pour la plage 60-80:", sum_60_to_80)
print("Somme des occurrences pour la plage 80-100:", sum_80_to_100)
```
